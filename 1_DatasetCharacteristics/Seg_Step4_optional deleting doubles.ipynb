```python
import os
import shutil

def delete_matching_files(folder1, folder2):
    # Get list of files in folder2
    files_in_folder2 = set(os.listdir(folder2))
    
    # Iterate through all files in folder1
    for filename in os.listdir(folder1):
        # Check if file in folder1 exists in folder2
        if filename in files_in_folder2:
            file_path = os.path.join(folder1, filename)
            
            # If it's a file (not a subdirectory), delete it
            if os.path.isfile(file_path):
                print(f"Deleting file: {file_path}")
                os.remove(file_path)

# Example usage
folder1 = 'marked_images' #here files get deleted
folder2 = 'second marked bb'

delete_matching_files(folder1, folder2)
```
