```python
import os
import re
import shutil

# Define source and target directories
source_folder = 'seamounts_seg'  # Change to your source folder
target_folder = 'seamounts_moved'  # Change to your target folder

# Regular expression to match the files with the expected PEAKID format
pattern = r'^\d+\.0\.png$'  # Matches files like '12345.0.png'

# Check if the target folder exists, create it if not
if not os.path.exists(target_folder):
    os.makedirs(target_folder)

# Loop through the files in the source folder
for filename in os.listdir(source_folder):
    # Check if the file is a .png and doesn't match the pattern
    if filename.endswith('.png') and not re.match(pattern, filename):
        # Construct the full file path
        file_path = os.path.join(source_folder, filename)
        
        # Move the file to the target folder
        shutil.move(file_path, os.path.join(target_folder, filename))
        print(f'Moved: {filename}')
```

```python
import os
import shutil

def move_files(source_folder, destination_folder):
    # Check if source and destination folders exist
    if not os.path.exists(source_folder):
        print(f"Source folder '{source_folder}' does not exist.")
        return
    if not os.path.exists(destination_folder):
        print(f"Destination folder '{destination_folder}' does not exist.")
        return

    # Get all files in the source folder
    for filename in os.listdir(source_folder):
        source_file = os.path.join(source_folder, filename)

        # Check if it's a file (not a directory)
        if os.path.isfile(source_file):
            destination_file = os.path.join(destination_folder, filename)
            
            # Move the file
            try:
                shutil.move(source_file, destination_file)
                print(f"Moved: {filename}")
            except Exception as e:
                print(f"Error moving {filename}: {e}")

# Example usage
source_folder = 'seamounts_moved_cropped'
destination_folder = 'seamounts_seg_cropped'

move_files(source_folder, destination_folder)
```

```python
# purge csv
import pandas as pd
import re

# Define the input file name (which will be overwritten)
file_name = 'merged_pixel_coordinates.csv'

# Read the CSV file into a pandas DataFrame
df = pd.read_csv(file_name)

# Define a regular expression pattern to match the valid image names (e.g., 123.0.png)
pattern = r'^\d+\.0\.png$'

# Filter the rows where the 'image_name' column matches the pattern
# Assuming that the column with image names is called 'image_name'
filtered_df = df[df['image_name'].apply(lambda x: bool(re.match(pattern, str(x))))]

# Overwrite the original CSV file with the filtered data
filtered_df.to_csv(file_name, index=False)

print(f"Original file '{file_name}' has been overwritten with filtered data.")
```
